```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(
    X, 
    y, 
    test_size=0.8, 
    random_state=42,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, 
    y_temp, 
    test_size=0.25, 
    random_state=42,
    stratify=y_temp
)
```

```python
# pro logistickou regresi neni dobre nechavat chybeici data na -1, pouziju imputaci na median

def replaceMissingValues(data, median, feature):
    data = data.replace(-1, np.nan).copy()
    data[feature] = data[feature].fillna(median)
    return data
    
X_train_imput = X_train.copy()
X_val_imput = X_val.copy()

for column in X_train_imput.columns:
    X_train_imput[column] = X_train_imput[column].replace(-1, np.nan).copy()
    X_val_imput[column] = X_val_imput[column].replace(-1, np.nan).copy()
        
    median = X_train_imput[column].median()
    X_train_imput = replaceMissingValues(X_train_imput, median, column)
    X_val_imput = replaceMissingValues(X_val_imput, median, column)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
def best_random_forest(x_data, y_data, x_val_data, y_val_data):
    best_estimators = 5
    best_roc = 0
    best_depth = 3
    
    for estimators in range (5, 400, 50):
        for depth in range (2,6,2):
            RFC_MODEL = RandomForestClassifier(
                n_estimators=estimators,
                max_depth=depth,
                class_weight='balanced',
                n_jobs=-1
            )
            
            RFC_MODEL.fit(x_data, y_data)
            y_val_proba_rfc = RFC_MODEL.predict_proba(x_val_data)[:, 1]
            roc_auc_rfc = roc_auc_score(y_val_data, y_val_proba_rfc)
    
    
            if roc_auc_rfc > best_roc:
                best_roc = roc_auc_rfc
                best_estimators = estimators
                best_depth = depth
    
            
    RFC_MODEL = RandomForestClassifier(
        n_estimators=best_estimators,
        max_depth=best_depth,
        class_weight='balanced',
        n_jobs=-1
    )
    
    RFC_MODEL.fit(x_data, y_data)
    y_val_proba_rfc = RFC_MODEL.predict_proba(x_val_data)[:, 1]
    roc_auc_rfc = roc_auc_score(y_val_data, y_val_proba_rfc)
    
    print(roc_auc_rfc)
    return RFC_MODEL
```

```python
best_random_forest(X_train_imput, y_train, X_val_imput, y_val)
```

```python

def best_decision_tree(x_data, y_data, x_val_data, y_val_data):
    max_roc = 0
    best_depth = 0
    
    for depth in range(1,10,2):
        DTC_MODEL = DecisionTreeClassifier(
        max_depth=depth,
        class_weight='balanced'
        )
        DTC_MODEL.fit(x_data, y_data)
        y_val_proba_imput = DTC_MODEL.predict_proba(x_val_data)[:, 1]
        roc_auc_imput_dtc = roc_auc_score(y_val_data, y_val_proba_imput)
        
        if roc_auc_imput_dtc > max_roc:
            best_depth = depth
            max_roc = roc_auc_imput_dtc
    
    DTC_MODEL = DecisionTreeClassifier(
        max_depth=best_depth,
        class_weight='balanced'
    )
    DTC_MODEL.fit(x_data, y_data)
    y_val_proba_imput = DTC_MODEL.predict_proba(x_val_data)[:, 1]
    roc_auc_imput_dtc = roc_auc_score(y_val_data, y_val_proba_imput)
    print(roc_auc_imput_dtc)

    return DTC_MODEL
```

```python
# pro logistickou regresi je potreba onehot encoding, pokud jsou data kategoricka neordinalni, nase sloupce jsou ordinalni, takze staci priznak ordinal



bin_uniform = KBinsDiscretizer(
    n_bins=5, 
    encode='ordinal',
    strategy='uniform', 
    subsample=None
)

bin_quantile = KBinsDiscretizer(
    n_bins=5, 
    encode='ordinal',
    strategy='quantile', 
    subsample=None
)
```

```python
def get_binned(bin_cols, X_data, X_val_data, discretizer_type, bins):
    

    binnarizer = KBinsDiscretizer(
        n_bins=bins, 
        encode='ordinal',
        strategy=discretizer_type, 
        subsample=None
    )
    
    X_train_bin = X_data.copy()
    X_val_bin = X_val_data.copy()


    for column in bin_cols:

        binnarizer.fit(X_train_bin[[column]])

        X_train_binned_data = binnarizer.transform(X_train_bin[[column]])
        X_val_binned_data = binnarizer.transform(X_val_bin[[column]])

        new_column_name = f'{column}_binned_ordinal'

        X_train_bin[new_column_name] = X_train_binned_data[:, 0]
        X_val_bin[new_column_name] = X_val_binned_data[:, 0]

        X_train_bin.drop(columns=[column], inplace=True)
        X_val_bin.drop(columns=[column], inplace=True)

    return X_train_bin, X_val_bin
```

```python
binning_cols = ["xcalc_13", "xcalc_14"]

X_train_BINNED, X_val_BINNED = get_binned(binning_cols, X_train_imput, X_val_imput, 'quantile', 6)
```

```python
# cviko 5
from sklearn.decomposition import PCA
```

pro PCA nemuzou byt data nenaskalovana a nemuzou zde byt kategoricka data

budu pca delat jen pro nekategoricke sloupce, dal by se udelat one-hot encoding, ale to by hodne nafouklo dataset,
lze to ale vyzkouset

```python
def getPCAValues(X_Train, X_Val, components):
    # ziskani nekategorickych sloupcu
    pca_cols = [col for col in X_Train.columns if col not in categorical_binary_cols]
    X_train_PCA_input = X_Train[pca_cols].copy()
    X_val_PCA_input = X_Val[pca_cols].copy()
    
    # skalovani nutne pro PCA
    scaler = StandardScaler()
    X_train_PCA_input = scaler.fit_transform(X_train_PCA_input)
    X_val_PCA_input = scaler.transform(X_val_PCA_input)

    
    # definice a trenink PCA
    pca = PCA(n_components=components, random_state=37)

    pca.fit(X_train_PCA_input)

    X_train_pca = pca.transform(X_train_PCA_input)
    X_val_pca = pca.transform(X_val_PCA_input)

    # rekonstrukce datasetu
    pca_component_names = [f'PCA_C_{i}' for i in range(1, pca.n_components_ + 1)]
    X_train_pca_df = pd.DataFrame(X_train_pca, columns=pca_component_names, index=X_Train.index)
    X_val_pca_df = pd.DataFrame(X_val_pca, columns=pca_component_names, index=X_Val.index)
   
    X_train_discrete = X_Train[categorical_binary_cols]
    X_val_discrete = X_Val[categorical_binary_cols]

    X_train_FINAL_PCA = pd.concat([X_train_pca_df, X_train_discrete], axis=1)
    X_val_FINAL_PCA = pd.concat([X_val_pca_df, X_val_discrete], axis=1)
    
    return X_train_FINAL_PCA, X_val_FINAL_PCA
```
